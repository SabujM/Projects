```python
#5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
#i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
r=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
htmlcontent=r.content
```

```python
soup=BeautifulSoup(htmlcontent,"html.parser")
```

```python
rank_team=[]
team=[]
matches=[]
points=[]
rating=[]
```

```python
block=soup.find_all("tr",{"class":"rankings-block__banner"})
for i in block:
    l=i.text.split()
    rank_team.append(l[0])
    team.append(l[1])
    matches.append(l[3])
    points.append(l[4])
    rating.append(l[5])
```

```python
others=soup.find_all("tr", {"class": "table-body"})
for i in others:
    l2=i.text.split()
    rank_team.append(l2[0])
    team.append(" ".join(l2[1:-4]))
    matches.append(l2[-3])
    points.append(l2[-2])
    rating.append(l2[-1])
```

```python
odi_top10_team=pd.DataFrame()
```

```python
odi_top10_team["Rank"]=rank_team[0:10]
odi_top10_team["Team"]=team[0:10]
odi_top10_team["Matches"]=matches[0:10]
odi_top10_team["Points"]=points[0:10]
odi_top10_team["Rating"]=rating[0:10]
```

```python
odi_top10_team
```

```python
odi_top10_team.to_csv("Top10 ODI Teams.csv")
```

```python
price=[]
prices=soup.find_all("span",{"class": "a-price-whole"})
for p in prices:
    price.append(p.text.strip())
    
len(price)
```

```python
len(image_links)
```

```python
ratings=[]
stars=soup.find_all("div", {"class":"a-row a-size-small"})
for i in stars:
    r=i.find("span").get("aria-label")
    ratings.append(r.split()[0])
len(ratings)
```

```python
url1="https://www.amazon.in/s?i=electronics&bbn=1389401031&rh=n%3A976419031%2Cn%3A1389401031%2Cn%3A1389432031%2Cp_36%3A-2000000&dc&page="
url2="qid=1610348912&rnid=1389401031&ref=sr_pg_"
for page in range(2, 267):
    req=requests.get(url1+str(page)+url2+str(page))
    contents=req.content
    soup=BeautifulSoup(contents,"html.parser")
    
  
    infos1=soup.find_all("a",{"class": "a-link-normal a-text-normal"})
    for i in infos1:
        product_name.append(i.text.strip())
    
    
    prices1=soup.find_all("span",{"class": "a-price-whole"})
    for p in prices1:
        price.append(p.text.strip())
        
    links=soup.find_all("img")
    for i in links:
        try:
            image_links.append(i.get("srcset").split()[0])
        except:
            pass

        
    stars1=soup.find_all("div", {"class":"a-row a-size-small"})
    for i in stars1:
        r=i.find("span").get("aria-label")
        ratings.append(r.split()[0])
    
```

```python
len(ratings), len(image_links), len(price), len (product_name)
```

```python
product_info["Product Name"]=product_name
product_info['Price (In IRS)']=price
product_info['Image Link']=image_links
product_info['Rating']=ratings
```

```python
#8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.
wether=requests.get("https://forecast.weather.gov/MapClick.php?w0=t&w7=rain&w8=thunder&w9=snow&w10=fzg&w11=sleet&AheadHour=0&Submit=Submit&FcstType=text&textField1=37.7771&textField2=-122.4196&site=all#.X_0tb14zbIV")
```

```python
weather_content=wether.content
```

```python
soup=BeautifulSoup(weather_content, "html.parser")
```

```python
period=[]
periods=soup.find_all("p", {"class":"period-name"})
for i in periods:
    period.append(i.text)
    
len(period)
```

```python
short_des=[]
divs=soup.find_all("div",{"class": "tombstone-container"})
for i in divs:
    short=i.find("p", {"class": "short-desc"})
    short_des.append(short.text)
    
len(short_des)
```

```python
low=[]
low_divs=soup.find_all("div",{"class": "tombstone-container"})
for i in low_divs:
    temp=i.find("p", {"class": "temp temp-low"})
    try:
        low.append(temp.text)
    except:
        low.append("-")
```

```python
high=[]
high_divs=soup.find_all("div",{"class": "tombstone-container"})
for i in high_divs:
    temp=i.find("p", {"class": "temp temp-high"})
    try:
        high.append(temp.text)
    except:
        high.append("-")
```

```python
des=[]
divs=soup.find_all("div",class_="col-sm-10 forecast-text")
for i in divs:
    des.append(i.text)
    
```

```python
len(period), len(short_des), len(low), len(high), len(des)
```

```python
import pandas as pd
weather=pd.DataFrame()
```

```python
weather["Period"]=period
weather["Short Description"]=short_des
weather["Low Temp."]=low
weather["High Temp."]=high
weather["Description"]=des[0:9]
```

```python
weather.to_csv("7 Days Weather Forecast, San Francisco.csv")
```
